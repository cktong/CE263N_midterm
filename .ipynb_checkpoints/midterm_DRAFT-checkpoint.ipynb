```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn

%matplotlib inline

file = open('housing_midterm_trn.csv', 'rb')
data = np.genfromtxt(file, delimiter=',',skip_header=1)

#print(data, data.shape)

x = data[:,0:20]
y = data[:,-1]
```

```python
c= ['longitude', 'latitude', 'yrbuilt', 'stories_code', 
    'TLA', 'Wall_code', 'beds', 'baths', 'halfbaths', 'frontage', 
    'garage_type_code', 'garage_sqft', 'rooms', 'lotsize', 'sale_date', 
    'sold93', 'sold94', 'sold95', 'sold96', 'sold97', 'sold98', 'price']


train_data= pd.DataFrame(data, columns = c)
```

```python
train_data.query('sold98 == 1').describe()
```

```python
# Importance of Location 

plt.figure(figsize=(18,12))

price = y
price[price>300000]= 300000

plt.scatter(x[:,0], x[:,1], s=10, c=price, alpha = 0.2, edgecolors='none')

plt.colorbar()
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

price = y
price[price>300000]= 300000

c= ['longitude', 'latitude', 'yrbuilt', 'stories_code', 
    'TLA', 'Wall_code', 'beds', 'baths', 'frontage', 
    'garage_type_code', 'garage_sqft', 'rooms', 'lotsize', 'sale_date']

features= train_data[c].as_matrix()

#features= train_data[['longitude', 'latitude','rooms', 'TLA','yrbuilt','price']].query('price <200000').as_matrix()
#price = list(features.transpose()).pop(-1)

pca = LinearDiscriminantAnalysis(n_components=2)
X_r = pca.fit(features,price).transform(features)

plt.figure(figsize=(18,12))
plt.scatter(X_r[:,0], X_r[:,1], s=10, c=price, alpha = 0.5, edgecolors='none')
plt.colorbar()

'''
xmin=-5
xmax=5
ymin= -2
ymax=2
plt.axis([xmin,xmax,ymin,ymax])
'''
```

```python
pca.predict()
```
